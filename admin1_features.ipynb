```python
# loading libraries 

import matplotlib as plt
import pandas as pd
import os
import pycountry
import string
from unidecode import unidecode
import category_encoders as ce

path = os.getcwd()
```

```python
# loading the data

adm1 = pd.read_csv(path + "/data/final_gdelt_bycountry.txt")
full = pd.read_csv(path + '/data/merged.csv')

# Dropping columns before 2000
adm1.drop(adm1.loc[adm1['year']<2000].index, inplace=True)
full.drop(full.loc[full['year']<2000].index, inplace=True)

# Filling missing UCDP data with 0
full[["deaths","state_deaths","nonstate_deaths","onesided_deaths","civilian_deaths","avg_sources","conflict_counts",
     "conflict_freq","dyad_counts","dyad_freq"]] = full[["deaths","state_deaths","nonstate_deaths","onesided_deaths",
                                                         "civilian_deaths","avg_sources","conflict_counts",
                                                         "conflict_freq","dyad_counts","dyad_freq"]].fillna(0)
```

```python
# Calculating the number of regions for each country in each of the Month/Year periods

number_regions = adm1.groupby(["isocode", "year", "month"]).nunique()["ActionGeo_ADM1Code"].reset_index()
number_regions = number_regions.rename({"ActionGeo_ADM1Code": "num_regions"}, axis=1)
number_regions.head(10)
```

```python
# Merging with the full dataset

full = pd.merge(full, number_regions, left_on=["isocode", "year", "month"], right_on = ["isocode", "year", "month"], 
                 how = "left")

full.head(10)
```

```python
# Dropping adm1 that does not have relevant events

adm1['fight'] = ((adm1['count_events_18'].fillna(0) > 0) | 
                        (adm1['count_events_19'].fillna(0) > 0) | 
                        (adm1['count_events_20'].fillna(0) > 0)).astype(int)

# The 'fight' column will contain 1 if there's a non-zero value in any of the columns, otherwise 0

adm1 = adm1[adm1['fight'] == 1]

# This will keep only the rows where 'fight' column is equal to 1
```

```python
# Merging the adm1 dataframe with the full dataframe

adm_full = pd.merge(full, adm1[['ActionGeo_ADM1Code',"isocode", "month", "year"]], 
                               left_on=["isocode", "month", "year"], right_on = ["isocode", "month", "year"], 
                               how = "left")

adm_full.head(10)
```

```python
# Target encoding the regions that have at least 1 event of type 18, 19 or 20

region_encoder = ce.TargetEncoder(smoothing = 1.0) 

# this can be done inside the loop if we want expanding window for the features

adm_full_train = adm_full[adm_full['year'] < 2014] # fix this once we establish the train years
adm_full_test = adm_full[adm_full['year'] > 2013] # fix this once we establish the train years

region_encoder.fit(adm_full_train['ActionGeo_ADM1Code'], adm_full_train['deaths'])

adm_full_train['ActionGeo_ADM1Code'] = region_encoder.transform(adm_full_train['ActionGeo_ADM1Code'], 
                                                               adm_full_train['deaths'])
adm_full_test['ActionGeo_ADM1Code'] = region_encoder.transform(adm_full_test['ActionGeo_ADM1Code'], 
                                                               adm_full_test['deaths'])

adm_full = pd.concat([adm_full_train, adm_full_test])
adm_full.head(10)
```

```python
# Getting the maximum, mean and median regions for each month/year and country

adm_full['Adm1_Max'] = adm_full.groupby(['isocode', 'month', 'year'])['ActionGeo_ADM1Code'].transform('max')
adm_full['Adm1_Mean'] = adm_full.groupby(['isocode', 'month', 'year'])['ActionGeo_ADM1Code'].transform('mean')
adm_full['Adm1_Median'] = adm_full.groupby(['isocode', 'month', 'year'])['ActionGeo_ADM1Code'].transform('median')

full_adm = adm_full.drop_duplicates(['isocode', 'month', 'year']).drop(["ActionGeo_ADM1Code"], 
                                                                      axis=1)
```

```python
# Creating a function for this:

import pandas as pd
import category_encoders as ce

def encode_adm(df, year_threshold):
    # Split the data into train and test based on the year threshold
    df_train = df[df['year'] < year_threshold]
    df_test = df[df['year'] >= year_threshold]

    # Target encoding the regions that have at least 1 event of type 18, 19, or 20
    region_encoder = ce.TargetEncoder(smoothing=1.0)
    region_encoder.fit(df_train['ActionGeo_ADM1Code'], df_train['deaths'])
    df_train['ActionGeo_ADM1Code'] = region_encoder.transform(df_train['ActionGeo_ADM1Code'], df_train['deaths'])
    df_test['ActionGeo_ADM1Code'] = region_encoder.transform(df_test['ActionGeo_ADM1Code'], df_test['deaths'])
    df = pd.concat([df_train, df_test])

    # Getting the maximum, mean, and median regions for each month/year and country
    df['Adm1_Max'] = df.groupby(['isocode', 'month', 'year'])['ActionGeo_ADM1Code'].transform('max')
    df['Adm1_Mean'] = df.groupby(['isocode', 'month', 'year'])['ActionGeo_ADM1Code'].transform('mean')
    df['Adm1_Median'] = df.groupby(['isocode', 'month', 'year'])['ActionGeo_ADM1Code'].transform('median')

    df_transform = df.drop_duplicates(['isocode', 'month', 'year']).drop(["ActionGeo_ADM1Code"], axis=1)

    return df_transform
```
