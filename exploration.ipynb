```python
# Concatenate Actor1Code and Actor2Code columns
merged_actors = pd.concat([df_explore["Actor1Code"], df_explore["Actor2Code"]])

# Calculate the number of unique values in the merged data
unique_actors = merged_actors.nunique()

print("Number of unique values in Actor1Code and Actor2Code:", unique_actors)
```

```python
import matplotlib.pyplot as plt
# Get the frequency of each EventRootCode
event_root_code_counts = df_explore["EventRootCode"].value_counts()

# Create a bar chart of the frequency
ax = event_root_code_counts.plot(kind="bar", figsize=(12, 6))

# Set the title and labels for the chart
ax.set_title("Frequency of Event Root Codes")
ax.set_xlabel("Event Root Code")
ax.set_ylabel("Frequency")

# Display the chart
plt.show()
```

```python
merged_geo = df_explore["Actor1Geo_FullName"].append(df_explore["Actor2Geo_FullName"]).reset_index(drop=True)


country_names = merged_geo.str.split(', ').str[-1]


country_counts = country_names.value_counts()


ax = country_counts.plot(kind="bar", figsize=(20, 8))


top_20_countries = country_counts.head(20)


ax = top_20_countries.plot(kind="bar", figsize=(20, 8))

ax.set_title("Number of Events by Country (Top 20)")
ax.set_xlabel("Country")
ax.set_ylabel("Number of Events")

# Display the chart
plt.show()
```

```python

combined_df = pd.concat([df_explore[['Actor1Geo_FullName', 'GLOBALEVENTID']], df_explore[['Actor2Geo_FullName', 'GLOBALEVENTID']].rename(columns={'Actor2Geo_FullName': 'Actor1Geo_FullName'})])


unique_event_df = combined_df.drop_duplicates(subset='GLOBALEVENTID', keep='first')


unique_event_df['Country'] = unique_event_df['Actor1Geo_FullName'].str.split(', ').str[-1]

country_counts = unique_event_df['Country'].value_counts()


top_20_countries = country_counts.head(20)
ax = top_20_countries.plot(kind="bar", figsize=(20, 8))
ax.set_title("Number of Unique GLOBALEVENTID's by Country (Top 20)")
ax.set_xlabel("Country")
ax.set_ylabel("Number of Unique GLOBALEVENTID's")
plt.show()
```

```python
import seaborn as sns
# Plot the KDE plot for the GoldsteinScale values
sns.kdeplot(df_explore['GoldsteinScale'], fill=True)

# Set the title and labels for the plot
plt.title('Distribution of GoldsteinScale Values')
plt.xlabel('GoldsteinScale Value')
plt.ylabel('Density')

# Display the plot
plt.show()
```
